```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import numpy as np

apps = pd.read_csv('datasets/apps.csv', dtype={'Installs':'str'})

apps.head()
```

```python
remove = [',', '+']

for re in remove:
    apps['Installs'] = apps['Installs'].apply(lambda x: x.replace(re,''))
    
apps.head()
```

```python
apps['Installs'] = apps['Installs'].astype(int)
```

```python
app_category_info = apps.groupby('Category').agg({'App':'count','Price':'mean','Rating':'mean'})
app_category_info.columns = ['Number of apps', 'Average price','Average rating']
app_category_info  
```

```python
top_apps = apps[(apps['Category'] == 'FINANCE') & (apps['Type'] == 'Free')]
```

```python
apps_review = pd.read_csv('datasets/user_reviews.csv')
apps_review.info()
```

```python
apps_and_review = top_apps.merge(apps_review, on = 'App')
apps_and_review
```

```python
feedback = apps_and_review.groupby('App').agg({'Sentiment Score':'mean'})

sort = feedback.sort_values('Sentiment Score', ascending=False)
sort['Sentiment Score']
```

```python
top_10_user_feedback = top_10 = sort.iloc[:10]
top_10_user_feedback
```
